# Garage and Cars (Multiply area by number for capacity)

```python
CarCapacity = ['GarageCars','GarageArea']
for var in CarCapacity:
    corr, _ = pearsonr(df_train[var], df_train['SalePrice'])
    print(var,':Pearsons correlation: %.3f' % corr)
```

```python
df_train['CarCapacity'] = df_train['GarageCars'] * df_train['GarageArea'] 
df_test['CarCapacity'] = df_test['GarageCars'] * df_test['GarageArea'] 
df_total['CarCapacity'] = df_total['GarageCars'] * df_total['GarageArea'] 
```

```python
corr, _ = pearsonr(df_train['CarCapacity'], df_train['SalePrice'])
print('Pearsons correlation: %.3f' % corr)
```

# Removing

```python
df_train = df_train.drop(['GarageCars','GarageArea'], axis = 1)
df_test = df_test.drop(['GarageCars','GarageArea' ], axis = 1)
df_total = df_total.drop(['GarageCars','GarageArea'], axis = 1)

df_train.to_csv('df_train_garage_car_clean.csv')
df_test.to_csv('df_test_garage_car_clean.csv')
df_total.to_csv('df_total_garage_car_clean.csv')
```
